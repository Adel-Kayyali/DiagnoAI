# **Splitting The DF**

**Separating the features & Target**

```python
X = df.drop(columns=['status'], axis=1)
Y = df['status']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data to training data & Test data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Models Training**

Support Vector Machine Model

```python
from sklearn import svm
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

**Random Forest Classifier Model**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth = 15, min_samples_leaf = 1, min_samples_split = 2, n_estimators = 100,random_state=47)
```

```python
rfc.fit(X_train, Y_train)
```

**Gradient Boosting Trees Model**

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbt = GradientBoostingClassifier(n_estimators=100, learning_rate=0.4, max_features=2, max_depth=11, random_state=47) 
```

```python
gbt.fit(X_train, Y_train)
```

# **Model Evaluation**

```python
from sklearn.metrics import accuracy_score
```

**Accuracy Score of SVC**

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of SVC Model : ', test_data_accuracy)
#Before Editing on the DF:    0.8717948717948718
#After Editing on the DF:     0.8205128205128205
```

**Accuracy Score of Random Forest Classifier**

```python
X_test_prediction = rfc.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of RFC Model : ', test_data_accuracy)
```

**Accuracy Score of Radient Boosting Trees**

```python
X_test_prediction = gbt.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of GBT Model : ', test_data_accuracy)
```
