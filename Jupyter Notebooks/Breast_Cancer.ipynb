# **Splitting the DF**

**Separating the features and target**

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the data into training data & Testing data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,stratify=Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Models Training**

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```


**Gradient Boosting Trees**

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
gbt = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_features=3, max_depth=10, random_state=47) 
```

```python
gbt.fit(X_train, Y_train)
```

# **Model Evaluation**

```python
from sklearn.metrics import accuracy_score
```

**Accuracy Score for Logestic Regression**

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on Logestic Regression model = ', test_data_accuracy)
```

**Accuracy Score for Gradient Boosting Trees Classifier**

```python
X_test_prediction = gbt.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on GBT = ', test_data_accuracy)

#100  - 0.956140350877193
#500  - 0.9649122807017544
#200  - 0.9736842105263158    Max depth=10    max_features=2
```
